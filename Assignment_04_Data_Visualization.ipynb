Use the “Titanic” dataset from Seaborn and do the following:
1. Countplot: Passenger Count by Embarked Port
● Create a countplot to display the number of passengers who boarded from each
port (embarked), separated by sex.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
print(df.head())
```

```python
sns.countplot(x='embarked',hue='sex',data=df)
plt.show()
```

Observation:
The majority of passengers embarked from 'S'.For all ports, there were generally more male passengers than female passengers.

2. Swarmplot: Age Distribution across Classes
● Create a swarmplot to visualize the distribution of age across passenger classes
(pclass), categorized by sex.

```python
sns.swarmplot(x='pclass',y='age',hue='sex',data=df)
plt.show()
```

Observation:Age distribution varies across passenger classes, with higher classes (1st) generally having older passengers.The density of points indicates the concentration of ages within each class and sex, showing some overlap but also distinct clusters.

3. Pie Chart: Gender Distribution
● Create a pie chart using Matplotlib to show the percentage of male vs. female
passengers. - Add percentage labels using autopct - Use explode for one slice to
highlight it.

```python
plt.pie(df['sex'].value_counts(),labels=df['sex'].unique(),autopct='%1.1f%%',explode=[0,0.1])
plt.show()
```

Observation:
The Titanic had more male passengers (64.8%) than female passengers (35.2%).

4. Heatmap: Correlation Matrix
● Create a heatmap to show correlations between age, fare, pclass, sibsp, parch,
and survived. (Use seaborn.heatmap with annotations)

```python
sns.heatmap(df[['age','fare','pclass','sibsp','parch','survived']].corr(),annot=True)
plt.show()
```

Observation:'survived' shows a negative correlation with 'pclass' (higher class, better survival chance) and 'sibsp'/'parch' (more siblings/spouses/children, lower survival chance).
'fare' and 'pclass' have a strong negative correlation, indicating higher fares for lower class numbers (1st class).

5. Violin Plot: Fare by Class and Gender
● Create a violin plot to visualize fare distribution across classes, separated by
gender.

```python
sns.violinplot(x='pclass',y='fare',hue='sex',data=df)
plt.show()
```

Observation:Fare distribution is significantly different across classes, with 1st class passengers paying much higher fares.
Within each class, there isn't a stark difference in fare distribution between male and female passengers, though 1st class females show a wider range of high fares.

6. Scatter Plot: Age vs Fare
● Create a scatter plot to explore the relationship between age and fare, colored by
survival status.
● Use a seaborn scatterplot with hue='survived'. Each plot must have titles, labels,
and legends where needed.

```python
sns.scatterplot(x='age',y='fare',hue='survived',data=df)
plt.show()
```

```python
sns.scatterplot(x='age',y='fare',hue='survived',data=df)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

Observation:
There doesn't appear to be a strong linear correlation between age and fare.
Passengers who survived (orange dots) are spread across various ages and fares, but there are noticeable clusters of survivors among higher fares, especially in the younger to middle-age groups.
