Account Login
● Allow users to log in using an account number/username and PIN/password
● Validate user credentials
● Display appropriate error messages for invalid login attempts

Amount Deposit
● Allow logged-in users to deposit money
● Deposit amount must be greater than zero
● Update and display the current balance after deposit

Amount Withdrawal
● Allow users to withdraw money
● Withdrawal amount must be greater than zero

● Withdrawal amount should not exceed the available balance
● Display the remaining balance after withdrawal

```python
class Account:
    def __init__(self, account_number, pin, balance=1000):
      self.account_number = account_number
      self.pin = pin
      self.__balance =balance
      self.login = False
    def accountlogin(self):
      while not self.login:
        account_number_input = int(input("Enter your account number: "))
        pin_input = int(input("Enter your PIN: "))
        if account_number_input == self.account_number and pin_input == self.pin:
          self.login = True
          print("Login successful!")
        else:
          print("Invalid account number or PIN. Please try again.")
    def depositamount(self):
      if self.login:
          amount = float(input("Enter the amount to deposit: "))
          if amount > 0:
              self.__balance += amount
              print(f"Deposit successful! Current balance: {self.__balance}")
          elif amount <= 0:
              print("Invalid amount. Please enter a positive value.")
      else:
          print("Please log in first.")
    def withdrawamount(self):
      if self.login:
        amount = float(input("Enter the amount to withdraw: "))
        if amount >0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawal successful! Current balance: {self.__balance}")
        elif amount <= 0:
          print("Invalid amount. Please enter a positive value.")
        elif amount > self.__balance:
          print("Insufficient balance. Withdrawal failed.")
      else:
        print("Please log in first.")


acc=Account(account_number=12345,pin=1234)
acc.accountlogin()
acc.depositamount()
acc.withdrawamount()
```
